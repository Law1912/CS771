```python
# Import libraries
import pickle
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Load the data
with open("mnist_small.pkl", "rb") as f:
    data = pickle.load(f)

# Extract the inputs and labels
X = data["X"]
Y = data["Y"].flatten()
colors = plt.cm.tab10(np.linspace(0, 1, 10))
```

```python
# Function to plot the projected data
def plot_data(X, Y, title):
    for i in range(10):
        plt.scatter(X[Y == i, 0], X[Y == i, 1], color=colors[i], label=str(i))
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
# Apply PCA to reduce the dimensionality to 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the PCA results
plot_data(X_pca, Y, "PCA on MNIST data")
```

```python
# Apply tSNE to reduce the dimensionality to 2
tsne = TSNE(n_components=2)
X_tsne = tsne.fit_transform(X)

# Plot the tSNE results
plot_data(X_tsne, Y, "tSNE on MNIST data")
```
